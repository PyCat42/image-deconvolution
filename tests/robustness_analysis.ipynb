# Robustness Analysis
_How do RL and NN algorithm behave for different degradations?_

```python
import pandas as pd
import torch
from torch.utils.data import DataLoader

from src.data_helpers import GeometricDataGenerator
from src.plotting_helpers import plot_comparative_results
from src.testing_helpers import comparative_testing
from src.NNdeconvolution import DeconvolutionUNet_Baseline
```

```python
BATCH_SIZE=32
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load best model
model = DeconvolutionUNet_Baseline()
state_dict = torch.load('nn_training/models/nn_baseline_model.pth', map_location=torch.device(device))
model.load_state_dict(state_dict)
model.eval()

model_about_df = pd.read_csv("nn_training/models/nn_baseline_model_about.csv")
```

```python
# Create full detector dataset with sigma=2.0
test_dataset_sigma2 = GeometricDataGenerator(
    num_samples=500,
    size=256,
    seed=42,
    sigma=2.0,
    poisson_noise=True,
    full_detector_model=True,
    data_loading=False
)
test_dataset_sigma2_loading = GeometricDataGenerator(
    num_samples=500,
    size=256,
    seed=42,
    sigma=2.0,
    poisson_noise=True,
    full_detector_model=True,
    data_loading=True
)
test_dataset_sigma2_dataloader = DataLoader(
    dataset=test_dataset_sigma2,
    batch_size=1
)
df_sigma2 = comparative_testing(
    dataset=test_dataset_sigma2,
    dataloader=test_dataset_sigma2_dataloader,
    model=model,
    model_about_df=model_about_df,
    device=device,
    path="comparison",
    filename="comparative_results_sigma2.csv"
)
plot_comparative_results(
    df=df_sigma2,
    save_name="comparative_results_sigma2"
)

# Create full detector dataset with sigma=5.0
test_dataset_sigma5 = GeometricDataGenerator(
    num_samples=500,
    size=256,
    seed=42,
    sigma=5.0,
    poisson_noise=True,
    full_detector_model=True,
    data_loading=False
)
test_dataset_sigma5_loading = GeometricDataGenerator(
    num_samples=500,
    size=256,
    seed=42,
    sigma=5.0,
    poisson_noise=True,
    full_detector_model=True,
    data_loading=True
)
test_dataset_sigma5_dataloader = DataLoader(
    dataset=test_dataset_sigma5,
    batch_size=1
)
df_sigma5 = comparative_testing(
    dataset=test_dataset_sigma5,
    dataloader=test_dataset_sigma5_dataloader,
    model=model,
    model_about_df=model_about_df,
    device=device,
    path="comparison",
    filename="comparative_results_sigma5.csv"
)
plot_comparative_results(
    df=df_sigma5,
    save_name="comparative_results_sigma5"
)

# Create full detector dataset with sigma=10.0
test_dataset_sigma10 = GeometricDataGenerator(
    num_samples=500,
    size=256,
    seed=42,
    sigma=10.0,
    poisson_noise=True,
    full_detector_model=True,
    data_loading=False
)
test_dataset_sigma10_loading = GeometricDataGenerator(
    num_samples=500,
    size=256,
    seed=42,
    sigma=10.0,
    poisson_noise=True,
    full_detector_model=True,
    data_loading=True
)
test_dataset_sigma10_dataloader = DataLoader(
    dataset=test_dataset_sigma10,
    batch_size=1
)

df_sigma10 = comparative_testing(
    dataset=test_dataset_sigma10,
    dataloader=test_dataset_sigma10_dataloader,
    model=model,
    model_about_df=model_about_df,
    device=device,
    path="comparison",
    filename="comparative_results_sigma10.csv"
)
plot_comparative_results(
    df=df_sigma10,
    save_name="comparative_results_sigma10"
)
```
